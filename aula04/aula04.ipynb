media

```python
def calcular_media(lista):
    return sum(lista) / len(lista)

numeros = [10,20,30,40,50]
media = calcular_media(numeros)
print(f'o media numero é {media}')
```

valor maximo

```python
def maximo(lista):
    return max(lista)

numeros = [1, 5, 7]
resultado = maximo(numeros)
print(resultado)
```

verificar todos os elementos de uma lista 

```python
def positivo_negativo(lista):
    positivos = [num for num in lista if num > 0]
    negativos = [num for num in lista if num < 0]
    return positivos, negativos

numeros = [10,-20,30,40,50,-3,7,8]

positivos, negativos = positivo_negativo(numeros)
print(positivos)
print(negativos)
```

retornar a soma dos elementos lista e a media, em variaveis separadas

```python
def soma_media(lista):
    soma = sum(lista)
    media = sum(lista)/len(lista)
    return soma, media

    
numeros = [10,2,7,9,10,22,5]

soma, media = soma_media(numeros)
print(soma)
print(f'{media:.2f}')
```

min e max

```python
def min_max(lista):
    minimo = min(lista)
    maximo = max(lista)
    return minimo, maximo

numeros = [10,2,7,9,10,22,5]

minimo, maximo = min_max(numeros)
print(minimo)
print(maximo)
```

retornar minimo, maximo, soma e media em uma unica função 

```python
def tudo(lista):
    minimo = min(lista)
    maximo = max(lista)
    soma = sum(lista)
    media = sum(lista)/len(lista)
    return minimo, maximo, soma, media

numeros = [10,2,7,9,10,22,5]

minimo, maximo, soma, media = tudo(numeros)
print(minimo)
print(maximo)
print(soma)
print(f'{media:.2f}')
    
```

calcular a soma os quadrado


```python
def quadrado(lista):
    return sum(num**2 for num in lista)

numeros = [10,2,7,9,10,22,5]

resultado = quadrado(numeros)
print(resultado)
```

contar quantas vezes um determinado elemento aparece na lista

```python
def contar_elemento(lista, elemento):
    return lista.count(elemento)

numeros = [1,2,3,4,5,1,2,3]
elemento = 3
ocorrencia = contar_elemento(numeros, elemento)
print(f'o elemento {elemento}, aparece {ocorrencia} vezes na lista')
```

inverter a lista 

```python
def inverter_lista(lista):
    return lista[::-1]

numeros = [1,2,3,4,5]
lista_invertida = inverter_lista(numeros)
print(f'lista invertida {lista_invertida}')
```

verificar se uma lista esta ordenada de forma crescente 

```python
def esta_ordenad_crescente(lista):
    return lista == sorted(lista)

numeros = [1,2,3,4,5]
esta_ordenada = esta_ordenad_crescente(numeros)
print(f'a lista esta ordenada de froma crescente {esta_ordenada}')
```

remova os elementos duplicados de uma lista

```python
def remover_duplicado(lista):
    return list(set(lista))

numeros = [1,2,3,4,5,1,2,3]
sem_duplicados = remover_duplicado(numeros)
print(f'lista sem elementos duplicados {sem_duplicados}')
```
