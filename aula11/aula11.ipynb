```python
import pandas as pd
```

```python
df = pd.read_csv('manutencao_preditiva.csv')
df
```

```python
# Selecionar os dados do Input e Output
 
# Inputs
 
x = df[['Temperatura Ar [K]',
        'Temperatura Processo [K]', 'Velocidade Rotacao [rpm]', 'Torque [Nm]',
        'Desgaste Ferramenta [min]']]
 
# Output
 
y = df['Tipo da Falha']
```

```python

from sklearn.model_selection import train_test_split
```

```python
x_treino, x_teste, y_treino, y_test = train_test_split(x, y, test_size = 0.3)
```

```python
from sklearn.neural_network import MLPClassifier

mlp = MLPClassifier(hidden_layer_sizes=(100), max_iter=100, activation='relu')
```

```python
# treinamento do modelo 

treino = mlp.fit(x_treino, y_treino)
```

```python
y_pred_treino = mlp.predict(x_treino)
y_pred_teste = mlp.predict(x_teste)
```
